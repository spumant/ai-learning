```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
#自己创建数据集
X, y = make_blobs(n_samples=500, n_features=2, centers=4, random_state=1)
```

```python
fig, ax1 = plt.subplots(1)
ax1.scatter(X[:, 0], X[:, 1]
            , marker='o'  #点的形状
            , s=8  #点的大小
            )
plt.show()
```

```python
color = ["red", "pink", "orange", "gray"]
fig, ax1 = plt.subplots(1)
for i in range(4):
    ax1.scatter(X[y == i, 0], X[y == i, 1]
                , marker='o'  #点的形状
                , s=8  #点的大小
                , c=color[i]
                )
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
n_clusters = 3
```

```python
cluster = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
```

```python
# 重要属性labels_，查看聚好的类别，每个样本所对应的类
y_pred = cluster.labels_
```

```python
y_pred
```

```python
# Kmeans因为并不需要建立模型或者预测结果，因此我们只需要fit就能得到聚类结果了
# Kmeans也有接口predict和fit_predict,表示学习数据X并对X的属性进行预测
# 但所得到的结果和我们不调用predict，直接fit后调用属性labels一模一样
pre = cluster.fit_predict(X)
```

```python
pre == y_pred
```

```python
# 我们什么时候需要predict呢？当数据量大的时候！
# 其实我们不必使用所有的数据来寻找质心，少量的数据就可以帮助我们确认质心了
# 当我们的数据量非常大的时候，我们可以使用部分数据来帮助我们确认质心
# 剩下的数据的聚类结果，使用predict来调用
```

```python
cluster_smallsub = KMeans(n_clusters=n_clusters, random_state=0).fit(X[:200])
```

```python
y_pred_ = cluster_smallsub.predict(X)
```

```python
y_pred == y_pred_
```

```python
# 但这样的结果，肯定与直接fit全部数据会不一致。有时候，当我们不需要那么精确，或者我们的数据量实在太大，那我们可以使用这种方法
```

```python
# 重要属性cluster_centers_，查看质心
centroid = cluster.cluster_centers_
```

```python
centroid
```

```python
centroid.shape
```

```python
# 重要属性inertia_，查看总距离的平方和
inertia = cluster.inertia_
inertia
```

```python
color = ["red", "pink", "orange", "gray"]
fig, ax1 = plt.subplots(1)
for i in range(n_clusters):
    ax1.scatter(X[y_pred == i, 0], X[y_pred == i, 1]
                , marker='o'
                , s=8
                , c=color[i]
                )
ax1.scatter(centroid[:, 0], centroid[:, 1]
            , marker="x"
            , s=15
            , c="black")
plt.show()
```

```python
n_clusters = 4
cluster_ = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
inertia_ = cluster_.inertia_
inertia_
```
